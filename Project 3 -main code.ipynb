## Defining functions ##

```python
def check_song (dataframe):
    user_song=input("Please write your choice here...")
    user_song=user_song.lower()
    trending_songs=list(pd.Series(dataframe["Song"]).str.lower())
    
    if user_song in trending_songs:
        return list(dataframe["Song"].sample())
    else:
        return user_song
```

```python
def playlist_extract(playlistid):
    
    pol_playlist=[]                     #In order to get rid of the limit we assign a new offset and iterate to extract all the songs of any playlist
    for i in range(1,1000,100):
        pol_playlist.append(spotify.user_playlist_tracks(user="spotify",playlist_id=playlistid,offset=i))
    

    pol_playlist_uri=[]                 #Iterate over items on the playlists to get the Uri of each song
    for item in pol_playlist:
        for song in item["items"]:
            pol_playlist_uri.append(song["track"]["uri"])

    pol_playlist_id=[]                  #Clean the Uri as some of them return errors
    for uri in pol_playlist_uri:
        if len(uri)<=36:
            pol_playlist_id.append(uri)
    
                                      
    pol_playlist_audiof=[]              #Iterate over each uri to get the audio features of each song
    for uri in pol_playlist_id:
        pol_playlist_audiof.append(spotify.audio_features(uri))
   

    audiofdf=pd.concat([pd.DataFrame(d) for d in pol_playlist_audiof])           #Create a dataframe with all the audio features of each song
    audiofdf.set_index("uri",inplace=True)
    audiofdf=audiofdf.reset_index()
    return audiofdf
```

```python
def audiofeat(song):
    
    results=spotify.search(q=song,limit=10)
    song_id=results["tracks"]["items"][0]["uri"]
    audiof=spotify.audio_features(song_id)
    
    return audiof
```

```python
def scale_userinput(audiof):
    audiof_df = pd.DataFrame(audiof)
    audiof_df = audiof_df.loc[:,['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms','time_signature']]
    audiof_scaled = scaler.transform(audiof_df)
    input_cluster = kmeans.predict(audiof_scaled)
    pd.Series(input_cluster).value_counts().sort_index()
    audiof_df["cluster"]=input_cluster
    return input_cluster
```

# Cheking elbow points #

```python
K = range(2, 20)

inertia = []
for n in K:
    kmeans = KMeans(n_clusters=n,
                    init="random", 
                    n_init=10,  
                    max_iter=300,
                    tol=0,
                    algorithm="full",
                    random_state=1234)
    kmeans.fit(final_df_scaled)
    inertia.append(kmeans.inertia_)
inertia 
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel("inertia")
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```
